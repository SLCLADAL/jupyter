Scores above 60 are considered plain language.<br><b>Flesch-Kincaid Grade</b> — US school grade level required to understand the text.<br><b>Gunning Fog Index</b> — years of formal education needed; below 12 is accessible.<br><b>SMOG Index</b> — similar to Fog; well-validated for health communication.<br><b>ARI (Automated Readability Index)</b> — character-based; works across genres.<br><b>Type-Token Ratio (TTR)</b> — lexical diversity; higher = more varied vocabulary.<br><b>Lexical Density</b> — proportion of content words (nouns, verbs, adjectives, adverbs).<br><b>Avg sentence length</b> — mean number of words per sentence.<br><b>Avg word length</b> — mean number of characters per word.</div>

````r
# ── CONFIGURATION ──────────────────────────────────────────────────
# Choose which readability measures to compute (TRUE/FALSE)
use_flesch        <- TRUE   # Flesch Reading Ease
use_fk_grade      <- TRUE   # Flesch-Kincaid Grade Level
use_fog           <- TRUE   # Gunning Fog Index
use_smog          <- TRUE   # SMOG Index
use_ari           <- TRUE   # Automated Readability Index
use_lexical_stats <- TRUE   # TTR, lexical density, sentence/word length
````

<div style="background:#51247a;color:white;padding:10px 18px;border-radius:6px;margin:1.4rem 0 .4rem 0;font-family:sans-serif;"><span style="font-size:.75rem;opacity:.7;text-transform:uppercase;letter-spacing:.05em;">Step 3</span><br><b style="font-size:.98rem;">&#x1F4CA; Run the analysis</b></div>


<div style="background:#f5f5f5;border-left:3px solid #bbb;border-radius:4px;padding:5px 12px;margin-bottom:3px;font-family:sans-serif;font-size:.78rem;color:#888;">&#x1F512; <em>Runs automatically — do not edit</em></div>


````r
suppressPackageStartupMessages(library(quanteda))
suppressPackageStartupMessages(library(quanteda.textstats))
suppressPackageStartupMessages(library(ggplot2))
suppressPackageStartupMessages(library(dplyr))
suppressPackageStartupMessages(library(tidyr))

seed_demo_texts("MyTexts")

texts <- tryCatch(load_texts("MyTexts"),
  error = function(e) { .err(conditionMessage(e)); NULL })

if (!is.null(texts)) {
  show_corpus_summary(texts)
  corp <- corpus(texts)
  .prog("Computing readability measures...", 30)
  # Collect selected formula names
  measures <- character(0)
  if (use_flesch)   measures <- c(measures, "Flesch")
  if (use_fk_grade) measures <- c(measures, "Flesch.Kincaid")
  if (use_fog)      measures <- c(measures, "FOG")
  if (use_smog)     measures <- c(measures, "SMOG")
  if (use_ari)      measures <- c(measures, "ARI")
  results_list <- list()
  if (length(measures) > 0) {
    rd <- textstat_readability(corp, measure = measures)
    results_list[["readability"]] <- as.data.frame(rd)
  }
  if (use_lexical_stats) {
    .prog("Computing lexical statistics...", 60)
    toks_all <- tokens(corp, remove_punct=TRUE, remove_numbers=FALSE)
    toks_lower <- tokens_tolower(toks_all)
    dfm_obj  <- dfm(toks_lower)
    # TTR per document
    ttr_vals <- ntype(dfm_obj) / ntoken(dfm_obj)
    # Avg sentence length
    sents <- lapply(as.list(corp), function(txt) {
      s <- unlist(strsplit(txt, "[.!?]+"))
      s <- s[nchar(trimws(s)) > 0]
      if (length(s)==0) return(NA_real_)
      mean(lengths(strsplit(trimws(s), "\\s+")))
    })
    avg_sent <- unlist(sents)
    # Avg word length
    avg_word <- sapply(as.list(toks_all), function(t) {
      w <- unlist(t); w <- w[nchar(w) > 0]
      if (length(w)==0) return(NA_real_)
      mean(nchar(w))
    })
    # Lexical density via POS tagging approximation
    # (content words = not in stopword list)
    sw <- stopwords("en")
    lex_density <- sapply(as.list(toks_lower), function(t) {
      w <- unlist(t); w <- w[nchar(w)>0]
      if (length(w)==0) return(NA_real_)
      round(sum(!w %in% sw) / length(w), 4)
    })
    lex_df <- data.frame(
      document       = names(texts),
      TTR            = round(ttr_vals, 4),
      LexDensity     = lex_density,
      AvgSentLength  = round(avg_sent, 2),
      AvgWordLength  = round(avg_word, 2),
      stringsAsFactors = FALSE
    )
    results_list[["lexical"]] <- lex_df
  }
  # Merge all results by document
  if (length(results_list) == 0) {
    .warn("No measures selected. Set at least one option to TRUE.")
  } else {
    combined <- Reduce(function(a, b) merge(a, b, by=1, all=TRUE), results_list)
    names(combined)[1] <- "Document"
    # Display results as HTML table
    col_names <- names(combined)
    th <- paste(sapply(col_names, function(h)
      paste0('<th style="padding:4px 10px;font-size:.8rem;color:#51247a;',
             'border-bottom:2px solid #d0c4e8;white-space:nowrap;">', h,
             "</th>")), collapse="")
    tr_rows <- apply(combined, 1, function(r)
      paste0("<tr>", paste(sapply(r, function(v)
        paste0('<td style="padding:3px 10px;font-size:.82rem;">',
               ifelse(is.na(v),"—",v), "</td>")), collapse=""), "</tr>"))
    IRdisplay::display_html(paste0(
      '<div style="font-family:sans-serif;margin:.5rem 0;">',
      '<b style="color:#51247a;font-size:.88rem;">Readability scores</b>',
      '<div style="overflow-x:auto;margin-top:6px;">',
      '<table style="border-collapse:collapse;">',
      '<thead style="background:#f4f0f8;"><tr>', th,
      '</tr></thead><tbody>',
      paste(tr_rows, collapse="\n"),
      "</tbody></table></div></div>"
    ))
    .prog("Plotting...", 85)
    # Plot available numeric readability measures
    plot_cols <- setdiff(names(combined),
                         c("Document","document","TTR","LexDensity",
                           "AvgSentLength","AvgWordLength"))
    if (length(plot_cols) > 0) {
      plot_df <- combined |>
        dplyr::select(Document, dplyr::all_of(plot_cols)) |>
        tidyr::pivot_longer(-Document, names_to="Measure", values_to="Score")
      p_rd <- ggplot(plot_df, aes(x=Document, y=Score, fill=Document)) +
        geom_col(show.legend=FALSE, width=.6) +
        facet_wrap(~Measure, scales="free_y") +
        scale_fill_viridis_d(option="plasma", begin=0.2, end=0.8) +
        theme_bw(base_size=11) +
        theme(axis.text.x=element_text(angle=30, hjust=1)) +
        labs(title="Readability scores by document",
             x=NULL, y="Score")
      print(p_rd)
    }
    # Plot lexical stats if computed
    if (use_lexical_stats && "TTR" %in% names(combined)) {
      lex_plot_df <- combined |>
        dplyr::select(Document,
                      dplyr::any_of(c("TTR","LexDensity","AvgSentLength","AvgWordLength"))) |>
        tidyr::pivot_longer(-Document, names_to="Measure", values_to="Value")
      p_lex <- ggplot(lex_plot_df, aes(x=Document, y=Value, fill=Document)) +
        geom_col(show.legend=FALSE, width=.6) +
        facet_wrap(~Measure, scales="free_y") +
        scale_fill_viridis_d(option="viridis", begin=0.2, end=0.8) +
        theme_bw(base_size=11) +
        theme(axis.text.x=element_text(angle=30, hjust=1)) +
        labs(title="Lexical statistics by document", x=NULL, y="Value")
      print(p_lex)
    }
    dir.create("MyOutput", showWarnings=FALSE, recursive=TRUE)
    save_excel(combined, "readability_scores.xlsx")
    out_files <- "readability_scores.xlsx"
    if (length(plot_cols) > 0) {
      ggsave("MyOutput/readability_plot.png", plot=p_rd,
             bg="white", width=9, height=6, dpi=200)
      out_files <- paste0(out_files, ", readability_plot.png")
    }
    if (use_lexical_stats && "TTR" %in% names(combined)) {
      ggsave("MyOutput/lexical_stats_plot.png", plot=p_lex,
             bg="white", width=9, height=6, dpi=200)
      out_files <- paste0(out_files, ", lexical_stats_plot.png")
    }
    .ok(paste0("Saved: ", out_files))
  }
}
````
