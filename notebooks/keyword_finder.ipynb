<div style="background:#51247a;color:white;padding:10px 18px;border-radius:6px;margin:1.4rem 0 .4rem 0;font-family:sans-serif;"><span style="font-size:.75rem;opacity:.7;text-transform:uppercase;letter-spacing:.05em;">Step 4</span><br><b style="font-size:.98rem;">&#x270F;&#xFE0F; Configure the analysis</b></div>


<div style="background:#e8f4fd;border-left:4px solid #4085C6;border-radius:4px;padding:7px 14px;margin-bottom:3px;font-family:sans-serif;font-size:.82rem;color:#2a5ea8;">&#x270F;&#xFE0F; <b>Edit the values below</b>, then run this cell (Shift+Enter or click &#x25B6; Run in the toolbar).</div>


````r
# ── CONFIGURATION ──────────────────────────────────────────────────
top_n           <- 15    # keywords to show per direction in the plot
min_freq        <- 2     # minimum word frequency to include
keyness_measure <- "G2"  # "G2" (log-likelihood), "chi2", or "lr" (log-ratio)
````

<div style="background:#51247a;color:white;padding:10px 18px;border-radius:6px;margin:1.4rem 0 .4rem 0;font-family:sans-serif;"><span style="font-size:.75rem;opacity:.7;text-transform:uppercase;letter-spacing:.05em;">Step 5</span><br><b style="font-size:.98rem;">&#x1F4CA; Run the analysis</b></div>


<div style="background:#f5f5f5;border-left:3px solid #bbb;border-radius:4px;padding:5px 12px;margin-bottom:3px;font-family:sans-serif;font-size:.78rem;color:#888;">&#x1F512; <em>Runs automatically — do not edit</em></div>


````r
suppressPackageStartupMessages(library(quanteda))
suppressPackageStartupMessages(library(quanteda.textstats))
suppressPackageStartupMessages(library(ggplot2))

# Seed demo data into Target and Reference if both folders are empty
dir.create("Target",    showWarnings=FALSE, recursive=TRUE)
dir.create("Reference", showWarnings=FALSE, recursive=TRUE)
if (length(list.files("Target",    pattern="(?i)\\.txt$")) == 0 &&
    length(list.files("Reference", pattern="(?i)\\.txt$")) == 0) {
  writeLines("The experiment was designed to test the hypothesis that exposure to ultraviolet radiation increases mutation rates in bacterial cultures. Samples were incubated under controlled conditions for forty-eight hours and subsequently analysed using spectrophotometric methods. Results indicated a statistically significant increase in mutation frequency among irradiated samples compared to controls. These findings are consistent with established models of DNA damage and repair. Further replication across multiple bacterial strains is recommended to confirm the generalisability of these results.", "Target/target_science_1.txt")
  writeLines("Cognitive load theory proposes that human working memory has a limited capacity and that instructional design should minimise unnecessary demands on this capacity. Empirical studies have demonstrated that split-attention effects, redundancy effects, and expertise reversal effects all influence learning outcomes in predictable ways. A meta-analysis of seventy-two studies found a moderate to large effect size for interventions based on cognitive load principles. Implications for educational technology design and curriculum development are discussed.", "Target/target_science_2.txt")
  writeLines("Local council voted last night to approve a new development plan for the waterfront area. The proposal, which has been under review for nearly two years, will see the construction of three hundred apartments and a new public park. Residents voiced mixed reactions at the meeting. Supporters argued the project would bring much-needed housing and boost the local economy. Critics expressed concern about traffic congestion and the loss of green space. The mayor said construction was expected to begin early next year pending final approvals.", "Reference/reference_news_1.txt")
  writeLines("The national football team secured a dramatic last-minute victory against their rivals on Saturday evening in front of a sold-out crowd. A penalty in the ninety-third minute decided the match after both sides had cancelled each other out for much of the game. The win puts the team top of their qualifying group with two games remaining. The coach praised the fighting spirit of his players and said he was confident they could secure qualification for the tournament with one more positive result.", "Reference/reference_news_2.txt")
  .info(paste0("Demo data loaded: 2 science texts in <b>Target</b>, ",
               "2 news texts in <b>Reference</b>. ",
               "Upload your own files and re-run to analyse your own corpus."))
}

target_texts <- tryCatch(load_texts("Target"),
  error = function(e) { .err(paste("Target folder:", conditionMessage(e))); NULL })
ref_texts <- tryCatch(load_texts("Reference"),
  error = function(e) { .err(paste("Reference folder:", conditionMessage(e))); NULL })

if (!is.null(target_texts) && !is.null(ref_texts)) {
  .info(paste0("Target: ", length(target_texts),
               " file(s) | Reference: ", length(ref_texts), " file(s)"))
  .prog("Building corpus and calculating keyness...", 40)
  all_texts <- c(target_texts, ref_texts)
  corp <- corpus(all_texts)
  docvars(corp, "group") <- c(rep("target",    length(target_texts)),
                              rep("reference", length(ref_texts)))
  dfm_all <- dfm(tokens(corp, remove_punct = TRUE) |> tokens_tolower()) |>
    dfm_trim(min_termfreq = min_freq)
  valid_measures <- c("G2", "chi2", "lr")
  if (!keyness_measure %in% valid_measures) {
    .warn(paste0("keyness_measure \"", keyness_measure, "\" not recognised. ",
                 "Using \"G2\". Valid: ", paste(valid_measures, collapse=", ")))
    keyness_measure <- "G2"
  }
  keys <- textstat_keyness(dfm_all,
           target  = docvars(corp, "group") == "target",
           measure = keyness_measure)
  if (nrow(keys) == 0) {
    .warn("No keywords found. Try reducing min_freq or uploading more text.")
  } else {
    keys$direction <- ifelse(keys[[keyness_measure]] > 0, "target", "reference")
    .prog("Plotting...", 80)
    top_target <- head(keys[keys$direction == "target",    ], top_n)
    top_ref    <- head(keys[keys$direction == "reference", ], top_n)
    plot_df    <- rbind(top_target, top_ref)
    measure_labels <- c(G2="G² (log-likelihood)", chi2="χ²", lr="Log-ratio")
    y_lab <- measure_labels[[keyness_measure]]
    p <- ggplot(plot_df, aes(x = reorder(feature, .data[[keyness_measure]]),
                             y = .data[[keyness_measure]], fill = direction)) +
      geom_col(width = .7) + coord_flip() +
      scale_fill_manual(values = c(target="#51247a", reference="#f0a500"),
                        labels = c(target="Target corpus", reference="Reference corpus"),
                        name = NULL) +
      theme_bw(base_size = 12) +
      theme(legend.position = "bottom") +
      labs(title = paste("Top", top_n, "keywords: target vs reference"),
           x = "Word", y = y_lab)
    print(p)
    save_excel(keys, "keywords.xlsx")
    dir.create("MyOutput", showWarnings=FALSE, recursive=TRUE)
    ggsave("MyOutput/keywords_plot.png", bg="white", width=8, height=5, dpi=200)
    .ok("Saved <b>keywords.xlsx</b> and <b>keywords_plot.png</b> to MyOutput.")
  }
}
````
